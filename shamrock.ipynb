```python
# Written by Aaron Shinn
# March 2024

import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt

def rotate(angle, x, y):
    x_r = x * np.cos(angle) + y * np.sin(angle)
    y_r = -x * np.sin(angle) + y * np.cos(angle)
    return x_r, y_r
    
def translate(x, y, reflect):
    delta_x = reflect*16  # adjusted for image fit; reflect parameter used to flip between left and right leaves
    delta_y = 22          # adjusted for image fit
    x_t = x + delta_x
    y_t = y - delta_y
    return x_t, y_t

# compute top leaf
t = np.arange(0, 2*np.pi, 0.1)
x1 = 16*(np.sin(t))**3
y1 = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

# transform for right leaf 
x2, y2 = rotate(0.6*np.pi, x1, y1)
x2, y2 = translate(x2, y2, 1.0) 

# transform for left leaf 
x3, y3 = rotate(-0.6*np.pi, x1, y1)
x3, y3 = translate(x3, y3, -1.0)

# add stem
x4 = np.arange(0.045, 10, 0.1)
y4 = (1 / x4) - 40    # adjusted for image fit

plt.figure(figsize=(12, 12))
plt.plot(x1,y1, x2, y2, x3, y3, x4, y4, color="green", linewidth = 10.0)
plt.fill(x1,y1,"green")
plt.fill(x2,y2,"green")
plt.fill(x3,y3,"green")
plt.show()
```
